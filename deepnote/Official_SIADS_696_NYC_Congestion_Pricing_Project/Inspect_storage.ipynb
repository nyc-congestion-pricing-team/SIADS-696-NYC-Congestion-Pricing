```
from pathlib import Path

total_bytes = 0

for f in Path("pipeline_data").rglob("*"):
    if f.is_file():
        total_bytes += f.stat().st_size

print(f"pipeline_data size: {total_bytes / 1024**3:.2f} GB")
```

```
from pathlib import Path

for f in sorted(Path("pipeline_data").rglob("*")):
    if f.is_file():
        print(
            f"{f.stat().st_size / 1024**2:10.2f} MB  {f}"
        )
```
